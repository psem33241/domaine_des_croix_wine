> 3.1 **Comp√©tence C1.4** Renvoyer le nom de la playlist qui contient le plus de morceaux, ainsi que le nombre de morceaux pour cette playlist.

Une r√©ponse rapide consiste en la requete suivante

```python
pd.read_sql("""
SELECT
    p.PlaylistId,
    p.Name AS nom_playlist,
    COUNT(pt.TrackId) AS nombre_morceaux
FROM
    Playlist p
INNER JOIN PlaylistTrack pt
    ON p.PlaylistId = pt.PlaylistId
GROUP BY
    p.PlaylistId,
    p.Name
ORDER BY
    COUNT(pt.TrackId) DESC
LIMIT 1
""", conn)
```

La requete ci-apr√®s affiche tous les resultats ex-aequo sans utiliser LIMIT c'est √† dire sans verification manuelle au prealable (donne directement une information en plus comme quoi la playlist 'Music' apparait 2 fois dans **Playlist** avec deux identifiants diff√©rents, pour avoir ce resultat avec la premeire requete il faudrait verifier manuellement au prealable et faire un LIMIT 2 ou 3 en fonction du resultat...)

```python
pd.read_sql("""
WITH PlaylistCounts AS (
    SELECT
        p.PlaylistId AS playlist_id,
        p.Name AS nom_playlist,
        COUNT(pt.TrackId) AS nombre_morceaux
    FROM
        Playlist p
    INNER JOIN PlaylistTrack pt
        ON p.PlaylistId = pt.PlaylistId
    GROUP BY
        p.PlaylistId,
        p.Name
),
MaxCount AS (
    SELECT MAX(nombre_morceaux) AS max_tracks
    FROM PlaylistCounts
)
SELECT
    pc.playlist_id,
    pc.nom_playlist,
    pc.nombre_morceaux
FROM
    PlaylistCounts pc, MaxCount mc
WHERE
    pc.nombre_morceaux = mc.max_tracks;
""", conn)
```

```python
pd.read_sql("""
SELECT
    a.Name AS Nom_artiste,
    COUNT(DISTINCT t.TrackId) AS Nombre_de_morceaux,
    COUNT(DISTINCT al.AlbumId) AS Nombre_albums
FROM
    Artist a
    INNER JOIN Album al ON a.ArtistId = al.ArtistId
    INNER JOIN Track t ON t.AlbumId = al.AlbumId
GROUP BY
    a.Name, a.ArtistId
ORDER BY
    Nombre_de_morceaux DESC;
""", conn)
```

## Regex

> 11. **Comp√©tence C2.7, Comp√©tence C2.3** Utiliser la librairie `re` (Regular Expression) pour extraire le code postal de l'adresse ci-dessous, √† l'aide d'une fonction que vous allez cr√©er:

```python
adresse = "8 Passage Petit Cerf 75017 Paris"
```

```python
import re

def extraire_code_postal(adresse):
    # Expression r√©guli√®re pour trouver un code postal fran√ßais (5 chiffres cons√©cutifs)
    pattern = r'\b\d{5}\b' # recherche dans une raw string 5 nombres consecutifs se trouvant √† une limite de mots
    match = re.search(pattern, adresse)
    if match:
        return match.group()
    else:
        return None # si pas de resultat, retourne None

# Application
code_postal = extraire_code_postal(adresse)
print("Code postal extrait :", code_postal)
```

Une pr√©cision de 1.0 signifie que toutes les pr√©dictions sont correctes, tandis qu'une pr√©cision de 0.0 signifie qu'aucune pr√©diction n'est correcte. De mani√®re g√©n√©rale, un score de pr√©cision √©lev√© indique que le mod√®le a bien appris √† partir des donn√©es d'entra√Ænement et qu'il est capable de faire des pr√©dictions pr√©cises sur les nouvelles donn√©es.

> 25. **Comp√©tence C3.6** On exploite ici des donn√©es pour le jeu d'entra√Ænement, et des donn√©es pour le jeu de test. Quelles sont les limites de cette m√©thode ?

**R√©ponse :**

L'utilisation de donn√©es s√©par√©es pour le jeu d'entra√Ænement et le jeu de test pr√©sente des limites, telles que :

- `La s√©lection de donn√©es non repr√©sentatives :`
  Si les donn√©es sont s√©lectionn√©es de mani√®re non repr√©sentative pour le jeu d'entra√Ænement ou le jeu de test, cela peut biaiser les r√©sultats et conduire √† des mod√®les over ou underfitt√©)

- `la limitation de la taille des √©chantillons :`
  Si le jeu de donn√©es est petit, la s√©paration en jeux d'entra√Ænement et de test r√©duit encore la quantit√© de donn√©es disponibles pour l'entra√Ænement, ce qui peut nuire √† la performance du mod√®le.

- `les difficult√©s li√©es √† la temporalit√© des donn√©es :`
  Dans certains cas, les donn√©es peuvent avoir une composante temporelle et si la division en jeux d'entra√Ænement et de test ne refl√®te pas correctement les changements au fil du temps alors cela represente un risque sur la performance du mod√®le.

Pour pallier ces limites et obtenir une √©valuation plus robuste, des m√©thodes comme la validation crois√©e peuvent √™tre utilis√©es, permettant d'exploiter pleinement l'ensemble des donn√©es disponibles et d'am√©liorer la g√©n√©ralisation du mod√®le.

> 26. Afficher la matrice de confusion (`confusion matrix`) et la commenter. Expliquer les m√©triques `precision` et `recall`.

**R√©ponse :**


```python
from sklearn.metrics import confusion_matrix

# Faire des pr√©dictions sur le jeu de test
y_pred = decision_tree_model.predict(X_test)

# Matrice de confusion
conf_matrix = confusion_matrix(y_test, y_pred)

# Affichage de la matrice de confusion avec seaborn
plt.figure(figsize=(10, 7))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='coolwarm')
plt.xlabel('Pr√©dictions')
plt.ylabel('V√©rit√©s')
plt.title('Matrice de confusion')
plt.show()
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

# Calculer l'exactitude
accuracy = accuracy_score(y_test, y_pred)
print(f"Exactitude : {accuracy:.2f}")

# Calculer la pr√©cision
precision = precision_score(y_test, y_pred, average='weighted')
print(f"Pr√©cision : {precision:.2f}")

# Calculer le rappel
recall = recall_score(y_test, y_pred, average='weighted')
print(f"Rappel : {recall:.2f}")

# Calculer le F1-score
f1 = f1_score(y_test, y_pred, average='weighted')
print(f"F1-score : {f1:.2f}")
```

La question pos√©e concerne Precision et Recall, nous rajoutons les m√©triques Exactitude et F1-score afin d'expliciter toutes les m√©triques d'autant plus qu'il y a souvent tr√®s grand econfision entre Accuracy et Precision (de par leur translation).

L'Exactitude (Accuracy) est la proportion de pr√©dictions correctes parmi toutes les pr√©dictions. Sa formule est (TP + TN)/ (TP+TN+FP+FN)

La Pr√©cision (Precision) est la proportion de pr√©dictions positives correctes parmi toutes les pr√©dictions positives. (TP)/(TP+FP)

Rappel (Recall) : La proportion de v√©ritables positifs correctement identifi√©s parmi tous les √©chantillons r√©ellement positifs. (TP)/(TP+FN)

F1-score : La moyenne harmonique de la pr√©cision et du rappel. 2*(Precision*Rappel)/(Precision+Rappel)

Cr√©ation d'une colonne cat√©gorielle bas√©e sur le prix

```python
# Analysons la distribution des prix pour d√©terminer les cat√©gories
plt.figure(figsize=(10, 6))
plt.hist(df['price'], bins=50)
plt.title('Distribution des prix')
plt.xlabel('Prix')
plt.ylabel('Fr√©quence')
plt.xscale('log')  # √âchelle logarithmique pour mieux visualiser
```

```python
# D√©finition des seuils pour les cat√©gories de prix
# Apr√®s analyse de la distribution, on peut d√©finir des cat√©gories pertinentes
def categorize_price(price):
    if price < 10:
        return 'Tr√®s bas'
    elif price < 50:
        return 'Bas'
    elif price < 100:
        return 'Moyen'
    elif price < 200:
        return '√âlev√©'
    elif price < 500:
        return 'Tr√®s √©lev√©'
    else:
        return 'Premium'

# Ajout de la nouvelle fonctionnalit√©
df['price_category'] = df['price'].apply(categorize_price)

# Ajout d'une fonctionnalit√© num√©rique pour les mod√®les
df['price_category_code'] = df['price_category'].map({
    'Tr√®s bas': 0,
    'Bas': 1,
    'Moyen': 2,
    '√âlev√©': 3,
    'Tr√®s √©lev√©': 4,
    'Premium': 5
})
```

```python
# Analyse des statistiques par cat√©gorie
category_stats = df.groupby('price_category')['price'].agg(['count', 'min', 'max', 'mean', 'median']).reset_index()
print("Statistiques par cat√©gorie de prix:")
category_stats
```

```python
# Visualisation de la r√©partition des cat√©gories de prix
plt.figure(figsize=(10, 6))
plt.bar(df['price_category'].value_counts().index, df['price_category'].value_counts().values)
plt.title('R√©partition des cat√©gories de prix')
plt.xlabel('Cat√©gorie')
plt.ylabel('Nombre d\'observations')
plt.xticks(rotation=45)
```

```python
# Conversion de la colonne 'price_category_code' en type objet
df['price_category_code'] = df['price_category_code'].astype(object)
```
